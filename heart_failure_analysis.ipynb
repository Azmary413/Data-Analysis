# Data Analysis

```python
df.high_blood_pressure.value_counts()
```

```python
df.high_blood_pressure.value_counts().plot(kind = "bar", color = ("orchid", "salmon"))
plt.title("high_blood_pressure")
plt.xlabel("0 = low risk     1 = high risk")
plt.ylabel("Amount");
```

```python
df.high_blood_pressure.value_counts().plot(kind = "pie", figsize = (9,6))
plt.legend(["low risk", "high risk"]);
```

## Which gender has the maximum heart_failure risk

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind = 'pie',figsize = (9,6))
plt.legend(["male", "female"]);
```

```python
pd.crosstab(df.sex, df.high_blood_pressure)
```

```python
sn.countplot(x = "high_blood_pressure", data = df, hue = "sex");
```

## Heart Failure risk for Anaemia patients

```python
pd.crosstab(df.anaemia, df.high_blood_pressure)
```

```python
pd.crosstab(df.anaemia, df.high_blood_pressure).plot(kind = "bar", color = ['coral', 'khaki'])
plt.title("Anaemia vs High blood pressure")
plt.xlabel("0 = No risk, 1 = Risk");
```

## Heart Failure risk for Diabetes patients

```python
pd.crosstab(df.diabetes, df.high_blood_pressure)
```

```python
pd.crosstab(df.diabetes, df.high_blood_pressure).plot(kind = "bar", color = ["orchid", "salmon"])
plt.title("Diabetes patients Heart failur Risk")
plt.xlabel("0 = No risk, 1 = Risk");
```

## Understanding Heart Failure risk in the elderly 

```python
sn.displot( x = "age", data = df, bins = 50, kde = True);
```

```python
plt.figure(figsize = (10,6))
plt.scatter(df.time[df.high_blood_pressure == 1],df.age[df.high_blood_pressure == 1],  c = "tomato")
plt.scatter(df.time[df.high_blood_pressure == 0],df.age[df.high_blood_pressure == 0],  c = "lightgreen")
plt.title('Heart Failure risk of aged people')
plt.xlabel("time")
plt.ylabel("age")
plt.legend(['Disease', 'No Disease']);
```

```python
sn.displot(x = 'time', data = df, bins = 30, kde = True, color = 'chocolate');
```

## Understanding Heart Failure risk for the ejection_fraction people

```python
plt.figure(figsize = (10,6))
plt.scatter(df.age[df.high_blood_pressure == 1],df.ejection_fraction[df.high_blood_pressure == 1],  c = "salmon")
plt.scatter(df.age[df.high_blood_pressure == 0],df.ejection_fraction[df.high_blood_pressure == 0],  c = "lightblue")
plt.title('Heart Failure risk of ejection_fraction people')
plt.xlabel("ejection_fraction")
plt.ylabel("age");
```

## Understanding Heart Failure risk for the smoking people

```python
sn.countplot(x = 'sex', data = df, hue = 'high_blood_pressure',palette="Set3")
plt.title('Heart Failure risk of smoking people');
```

```python
sn.displot(x = 'time', data = df, bins = 30, kde = True, color = 'chocolate');
```

## Death for High blood pressure 

```python
pd.crosstab(df.age[df.high_blood_pressure == 1], df.DEATH_EVENT[df.high_blood_pressure == 1]).plot(kind = "bar", color = ['coral', 'lightgreen'])
plt.title("DEATH_EVENT vs High blood pressure")
plt.xlabel("0 = Alive,    1 = DEATH");
```

```python
sn.displot(x = df.age[df.high_blood_pressure == 1], data = df, bins = 30, kde = True, color = 'coral');
plt.title("No. of oldest people with heart disease")
plt.xlabel("0 = low risk     1 = high risk")
plt.ylabel("Amount");
```

```python
sn.displot(x = df.age[df.high_blood_pressure == 0], data = df, bins = 30, kde = True, color = 'lightgreen');
plt.title("No. of oldest people without heart disease")
```

```python
sn.displot(x = 'creatinine_phosphokinase', data = df, bins = 30, kde = True, color = 'teal');
plt.title('creatinine_phosphokinase');
```

```python
sn.kdeplot(x = 'age', y = 'creatinine_phosphokinase', data = df, color = 'olive');
```
